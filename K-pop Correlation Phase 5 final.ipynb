## Why was this dataset created?

### Korean Tourism Dataset: 
The dataset was created as a foundation for amplification of the Korean tourism market. It also lays the framework for formulating future marketing strategies, thus allowing Korea to expand and tailor the tourism experience towards more targeted individuals. 

The dataset identifies the total number of visitors that came into South Korea from 2016- 2023. It also contains the purpose of visit(tourism, business, education, etc.), mode of transportation, gender, and whether the visitors were flight attendants or not (to differentiate from “real” visitors to flight attendants). 

### Reddit Dataset:
The dataset was created to maintain a record of officially announced K-pop events worldwide from 2016-2023, such as concerts, fanmeets, and festivals, specifically for the /r/kpop subreddit community. It serves as an archive for reference and planning purposes for K-pop fans who use Reddit.

### Youtube Dataset
The dataset was created to organize the top trending videos on Youtube each day, as an updated version of the Youtube Trending Video Statistics and Comments dataset. It is a daily record of the top trending videos, which Youtube identifies through a combination of factors including a measure of user interactions (e.g. likes, views, comments).

The Korean government has a separate entity called “Visit Korea Datalab” (by the Korean Tourism Organization) which focuses on the research of Korean tourism. - Identify and address multicollinearity among the independent variables. Including checking correlation matrices, removing or combining highly correlated variables, and using regularization techniques that penalize large coefficients.

The insignificance of the 'Event Count' coefficient raises questions about the direct impact of K-pop events on tourism within the constraints of this model, hinting that the model may not be capturing the complex dynamics of cultural influence on travel behavior.

Multicollinearity, as suggested by the moderate condition number, also poses a problem, potentially leading to inflated variances of the estimated coefficients and making them unreliable. The lack of statistical significance in the 'Event Count' variable further complicates the interpretation, as it implies that the presence of K-pop events doesn't significantly affect the tourism numbers, based on the current model structure. This could either reflect reality or indicate that the model is missing key variables that could provide a more comprehensive understanding of what drives tourism to South Korea.

To address the issue of multicollinearity and identify the variables contributing to the problem, we plan to calculate variance inflation factors (VIFs) for each predictor variable. By doing so, we can determine if any variables have high VIFs and may need to be removed or combined with others to reduce redundancy. Additionally, for insignificant variables like 'Event Count', we should consider exploring different model structures or non-linear relationships that may better capture the dynamics between K-pop events and tourism. Finally, the regression analysis appears to be further limited by potential serial correlation and non-normal residuals, both of which violate the assumptions of OLS and can lead to biased inference. Without checks for the normality of residuals and adjustments for any serial correlation, the model's predictive power and the reliability of its hypothesis tests may be compromised.
